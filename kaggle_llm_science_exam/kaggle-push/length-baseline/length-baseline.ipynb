# LLM Science Length Baseline

Ranks answer options by descending word count and writes `submission.csv`.


```python
import csv
import re
from pathlib import Path

OUTPUT_PATH = Path("/kaggle/working/submission.csv")
OPTIONS = ("A", "B", "C", "D", "E")
WORD_RE = re.compile(r"[A-Za-z0-9]+(?:[-'][A-Za-z0-9]+)?")


def find_test_csv():
    candidates = sorted(Path("/kaggle/input").glob("**/test.csv"))
    if not candidates:
        raise FileNotFoundError("No test.csv found under /kaggle/input")
    return candidates[0]


def word_count(text):
    return len(WORD_RE.findall(text or ""))


input_path = find_test_csv()
print(f"Using input: {input_path}")

with input_path.open(newline="", encoding="utf-8") as src, OUTPUT_PATH.open(
    "w", newline="", encoding="utf-8"
) as dst:
    reader = csv.DictReader(src)
    writer = csv.DictWriter(dst, fieldnames=["id", "prediction"])
    writer.writeheader()

    for row in reader:
        ranked = sorted(OPTIONS, key=lambda option: (-word_count(row[option]), option))
        writer.writerow({"id": row["id"], "prediction": " ".join(ranked[:3])})

print(f"Wrote {OUTPUT_PATH}")
```
